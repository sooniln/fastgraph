```python
import matplotlib.pyplot as plt
import numpy as np

# make data:
data_guava = [
    ("GuavaNetwork", 1143.011008),
    ("GuavaGraph", 792.563272),
    ("GuavaSimpleGraph", 683.802456),
]
data_jgraph = [
    ("JGraphTNetwork", 1579.150672),
    ("JGraphTGraph", 1579.150656),
    ("JGraphTSimpleGraph", 1579.150656),
]
data_fastgraph_topology = [
    ("MutableNetwork", 356.343072),
    ("MutableGraph", 120.385248),
    ("ImmutableGraph", 58.082984),
    ("ImmutableNetwork", 167.848728),
    ("MutableSimpleGraph", 120.385248),
    ("ImmutableSimpleGraph", 58.082984),
]
data_fastgraph_data = [
    ("MutableNetwork", 37.880920),
    ("MutableGraph", 202.287200),
    ("ImmutableGraph", 201.999208),
    ("ImmutableNetwork", 27.862760),
    ("MutableSimpleGraph", .960472),
    ("ImmutableSimpleGraph", .672504),
]

plt.figure(figsize=(8, 5))
plt.title("Memory Usage")
plt.ylabel("Size (MB)")
plt.xticks(rotation=90)
plt.bar(
    [name for name, _ in data_jgraph],
    [size for _, size in data_jgraph],
    capsize=5,
    color = 'g',
    label = "JGraphT")
plt.bar(
    [name for name, _ in data_guava],
    [size for _, size in data_guava],
    capsize=5,
    color = 'b',
    label = "Guava Graphs")
plt.bar(
    [name for name, _ in data_fastgraph_topology],
    [size for _, size in data_fastgraph_topology],
    capsize=5,
    color = 'r',
    label = "FastGraph Topology",)
plt.bar(
    [name for name, _ in data_fastgraph_data],
    [size for _, size in data_fastgraph_data],
    bottom=[size for _, size in data_fastgraph_topology],
    capsize=5,
    color = 'm',
    label = "FastGraph Data",)
plt.legend()
plt.show()
```
